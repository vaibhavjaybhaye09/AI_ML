```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
plt.style.use('seaborn-v0_8')
plt.rcParams['figure.figsize'] =(12,6)
plt.rcParams['figure.dpi'] = 100
```

```
np.random.seed(42)
```

```
np_students = 1000
```

```
df = pd.DataFrame({
    'students_id':range(1, np_students + 1),
    'age':np.random.randint(15,23,np_students),
    'gender': np.random.choice(['Male','Female'],np_students),
    'study_time':np.clip(np.random.normal(10,4,np_students),0,30),
    'attendance':np.clip(np.random.normal(85,15,np_students),0,100),
    'internet_access':np.random.choice(['YES','NO'],np_students),
    'parent_education':np.random.choice(['High School', 'some college','Bachelor','Master','PhD'],
                                      np_students,
                                      p = [0.2,0.25,0.3,0.2,0.05]),
    'family_support': np.random.choice(['YES','NO'],np_students,
                                       p = [0.70,0.30])

})
```
