```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
plt.style.use('seaborn-v0_8')
plt.rcParams['figure.figsize'] =(12,6)
plt.rcParams['figure.dpi'] = 100
```

```
np.random.seed(42)
```

```
np_students = 1000
```

```
df = pd.DataFrame({
    'students_id':range(1, np_students + 1),
    'age':np.random.randint(15,23,np_students),
    'gender': np.random.choice(['Male','Female'],np_students),
    'study_time':np.clip(np.random.normal(10,4,np_students),0,30),
    'attendance':np.clip(np.random.normal(85,15,np_students),0,100),
    'internet_access':np.random.choice(['YES','NO'],np_students),
    'parent_education':np.random.choice(['High School', 'some college','Bachelor','Master','PhD'],
                                      np_students,
                                      p = [0.2,0.25,0.3,0.2,0.05]),
    'family_support': np.random.choice(['YES','NO'],np_students,
                                       p = [0.70,0.30]),
    'extracurricular_activities': np.random.choice(['YES','NO'],
                                                   np_students,
                                                   p = [0.70,0.30]),
    'math_score': np.clip(np.random.normal(70,15,np_students)+
                         np.random.choice([0,5,10],np_students, p =[0.3,0.5,0.2]),0,100),
     'reading_score':np.clip(
        np.random.normal(72,14,np_students) +
        np.random.choice([0,5,10],np_students, p=[0.3,0.5,0.2]),
        0,100
                           ),
    'writing_score':np.clip(
        np.random.normal(71,15,np_students) +
        np.random.choice([0,5,10],np_students,p=[0.3,0.5,0.2]),
        0,100
    )


})
```

```
display(df.describe)
```

```
display(df.describe())
```

```
df['average_score'] = df[['math_score','reading_score','writing_score']].mean(axis=1)
df['average_score']
```

```
fig , ax = plt.subplots(figsize=(10,6))
ax.hist(df['average_score'],bins=20,color='steelblue',edgecolor='black', )
ax.axvline(df['average_score'].mean(),color='red',linestyle='dashed',linewidth=2)
ax.axvline(df['average_score'].median(),color='g',linestyle='dotted',linewidth=2)
ax.set_xlabel('Average Score')
ax.set_ylabel('number of students (Frequency)')
ax.set_title('Distribution of Average Scores')
ax.grid(True,)
```
