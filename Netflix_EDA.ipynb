```python
def rating_group(rating):
  if rating == 'TV-MA' or rating == 'R':
      return 'Adults'
  elif rating == 'PG-13' or rating == 'PG':
      return 'Teens'
  elif rating == 'TV-14' or rating == 'TV-PG':
      return 'Young & Adults'
  elif rating == 'NR':
      return 'N/A'
  else:
    return 'Children'
```

```python
df['Ratings'] = df['rating'].apply(rating_group)
```
