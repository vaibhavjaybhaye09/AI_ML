```
def size_pocess(item):
  if str(item)[-1] == "M":
    res = float(str(item).replace("M",""))
    res = res*1024
    return res
  elif str(item)[-1] == "k":
    res = float(str(item).replace("k",""))
    return res
  else :
    return str(np.nan)
```

```
df_copy['Size'] = df_copy['Size'].apply(size_pocess)
```
